```python
from datasets import load_dataset


# Split the data into 20% test, 80% training

data = load_dataset("chloecchng/biomedical_cpgQA", split='train')
data = data.train_test_split(test_size=0.2)
```

```python
data['train'].to_csv('cpgQA-train.csv')
```

```python
test_data = data['test']
test_n_val = test_data.train_test_split(test_size=0.5)
```

```python
test_n_val['test'].to_csv('cpgQA-test.csv')
test_n_val['train'].to_csv('cpgQA-validation.csv')
```
